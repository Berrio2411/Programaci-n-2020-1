#Punto 1

```
pesosColombianos= None
while(pesosColombianos==None):
  try:
    pesosColombianos= float(input('Ingrese su valor en pesos colombianos : '))
  except ValueError:
    print('El valor ingresado no es valido')
dolares=pesosColombianos/3800
print(f'el valor que usted ingreso en pesos equivale a {dolares} dolares')
```

#Punto 3

```
import matplotlib.pyplot as plt

meses =['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre' ]
ingresos =[1000, 900,1100,3000,3500,6000,4000,3400,7000,8000,5500,10000]

plt.bar(meses,ingresos)
plt.title('Meses vs Ingresos')
plt.xlabel('Meses del año')
plt.ylabel('Ingresos mensuales')
plt.savefig('GraficoIngresos.png')
plt.show()
```

```
import matplotlib.pyplot as plt
ciudades =['Bogota','Medellin','Barranquilla','Cali','Cartagena']
area=[1775,382,154,619 ,609]
explodeList=[0.20,0,0,0,0]
plt.title('Ciudades de Colombia segun su área en Km2')
plt.pie(area, labels=ciudades ,shadow=True , explode=explodeList,startangle=45)
plt.savefig('ciudades.png')
plt.show()
```

```
import pandas as pd
ecg = pd.read_csv('ecg (1).csv',encoding='UTF-8', header=0, delimiter=';').to_dict()
ecg.keys()
x = list (ecg['muestra'].values())
y = list (ecg['valor'].values())
import matplotlib.pyplot as plt
plt.plot(x,y,)
plt.title('Grafica ECG')
plt.xlabel('Tiempo(ms)')
plt.ylabel ('Voltaje (uV)')
plt.savefig('ecg.png')
plt.show()
```
